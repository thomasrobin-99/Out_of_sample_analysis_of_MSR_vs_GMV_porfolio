```python
import requests
# get the list
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'}
response = requests.get(url, headers=headers)
tables = pd.read_html(response.text)
sp500 = tables[1]

sp500_tickers = sp500['Symbol'].tolist()

start_date = '2000-01-01'
end_date = '2022-12-31'

# stock price
sp500_prices = yf.download(
    tickers=sp500_tickers,
    start=start_date,
    end=end_date,
    auto_adjust=False)['Adj Close']

# data cleaning
sp500_prices = sp500_prices.dropna(axis=1, how='all')

missing_counts = sp500_prices.isna().sum(axis=1) 

sp500_prices = sp500_prices[missing_counts <= 100]

sp500_prices = sp500_prices.loc[:, sp500_prices.isna().sum() <= 100]

sp500_prices = sp500_prices.reset_index('Date') 

sp500_prices
```

# Q 2.2

```python

# Compute daily log-returns

prices = sp500_prices.set_index('Date').sort_index()

log_returns = np.log(prices / prices.shift(1)).dropna()
```

# Q 2.3

```python
# Parameters
T = 252           
rf = 0.0           

# Annualized return
annualized_return = log_returns.mean() * T

# Annualized volatility
annualized_vol = log_returns.std() * np.sqrt(T)

# Annualized Sharpe ratio

excess_return = annualized_return - rf
annualized_sharpe = excess_return / annualized_vol


print("Annualized return (log-returns) :")
print(annualized_return.sort_values(ascending=False))

print("\n Annualized volatility :")
print(annualized_vol.sort_values())

print("\n Annualized Sharpe ratio :")
print(annualized_sharpe.sort_values(ascending=False))
```

```python
# No, it would not make sense to choose portfolio weights based only on 
# the Sharpe ratios of the individual stocks.
#
# The Sharpe ratio evaluates each asset in isolation and ignores the 
# covariances between assets, which are essential in portfolio construction. 
# As shown in the course, optimal mean-variance weights depend on the entire 
# covariance matrix V[R], not just individual risk–return profiles.
#
# In addition, Sharpe ratios rely on estimates of expected returns, which are 
# very imprecise and lead to poor out-of-sample performance.
#
# Selecting assets purely by Sharpe ratios also produces concentrated and 
# poorly diversified portfolios.
#
# Conclusion: the Sharpe ratio is useful for descriptive comparison, 
# but it should not be used alone to determine portfolio weights.
```

```python
# solve function
import cvxpy as cp

def solve_mvp(mu, sigma):
    mu = np.array(mu).flatten()          # ensure 1D array
    sigma = np.array(sigma)              # ensure matrix

    n = len(mu)
    w = cp.Variable(n)

    objective = cp.Maximize(mu @ w)
    constraints = [
        cp.sum(w) == 1,
        w >= 0
    ]

    prob = cp.Problem(objective, constraints)
    prob.solve()

    return w.value


def solve_gmv(sigma):
    sigma = np.array(sigma)
    n = sigma.shape[0]

    w = cp.Variable(n)

    objective = cp.Minimize(cp.quad_form(w, sigma))
    constraints = [
        cp.sum(w) == 1,
        w >= 0
    ]

    prob = cp.Problem(objective, constraints)
    prob.solve()

    return w.value

rf = .0
mu = returns_m_60.mean()
excess_mu = mu - rf
sigma = returns_m_60.cov()

w_mvp = solve_mvp(excess_mu, sigma)
w_gmv = solve_gmv(sigma)

w_gmv
```

```python
# minimize function

# MV
rf = .0
mu = returns_m_60.mean()
excess_mu = mu - rf
sigma = returns_m_60.cov()
n = len(mu)


from scipy.optimize import minimize

w_mv_c = np.ones(n) / n
constraints = {'type': 'eq', 'fun': lambda w: np.sum(w) - 1}
bounds = [(0, 1) for _ in range(n)]

def sharpe_ratio(w, mu, cov_matrix, rf):
    port_return = w.T @ mu
    port_vol = np.sqrt(w.T @ cov_matrix @ w)
    return -(port_return - rf) / port_vol

result = minimize(sharpe_ratio, 
                  w_mv_c, 
                  args=(mu, sigma, rf),
                  method='SLSQP', 
                  bounds=bounds, 
                  constraints=constraints)

w_mv = result.x
print(w_mv)
```

```python
# minimize function

# GMV
mu = returns_m_60.mean()
sigma = returns_m_60.cov()
n = len(mu)
ones = np.ones(len(sigma))

w_gmv_c = np.ones(n) / n
constraints = {'type': 'eq', 'fun': lambda w: np.sum(w) - 1}
bounds = [(0, 1) for _ in range(n)]

def portfolio_variance(w, cov_matrix): 
    return w.T @ cov_matrix @ w

result = minimize(portfolio_variance, 
                  w_gmv_c, 
                  args=(sigma,),
                  method='SLSQP', 
                  bounds=bounds, 
                  constraints=constraints)

w_gmv = result.x
print(w_gmv)
```
